```python
tmdb_movies = tmdb_movies.sort_values(by= "revenue", ascending= False).iloc[:5]
tmdb_movies
```

```python
ax = sns.barplot(x= "revenue", y = "genres", data =tmdb_movies)

#name of the title
plt.title("Revenue vs Genres")

#Add axis titles
plt.xlabel("Revenue")
plt.ylabel("Genre")
```

the graph above shows how various genres are doing in terms of their revenues

```python
ax = sns.barplot(x= "release_year", y = "revenue", data =tmdb_movies)

#name of the title
plt.title("Release year vs Revenue")

#Add axis titles
plt.xlabel("Release year")
plt.ylabel("Revenue")
```

the code displays the graph of whether the revenue brought by the movie was dependant on the year the movie was relesead

```python
ax = sns.barplot(x= "director", y = "revenue", data =tmdb_movies)

#name of the title
plt.title("revenue brought by the directors")

#Add axis titles
plt.xlabel("Director")
plt.ylabel("Revenue")
```

the graph above shows the various directors and how much revenue they were able to bring from the movies

```python
ax = sns.barplot(x= "revenue", y = "production_companies",data =tmdb_movies)

#name of the title
plt.title("Revenue vs Production companies")

#Add axis titles
plt.xlabel("Revenue")
plt.ylabel("Popularity companies")
```

this graph shows whether the revenue of the film is based on the production companies
